# 1. IMPORTING LIBRARIES AND DATASET

```python
import pandas as pd #----for data manipulation
import numpy as np #---data statical analysis
import matplotlib.pyplot as plt #data visualziation
import seaborn as sns #statical visualization

from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV

from sklearn.svm import SVC
from sklearn.metrics import classification_report,confusion_matrix
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
#loading our data set 
breastcancer_df = load_breast_cancer()
```

```python
print(breastcancer_df['feature_names'])
```

```python
breastcancer_df['data'].shape #569 data samples with 30 feature columns each
```

```python
#our data inside cancer is not easily readable so we use pandas to include all the
#569 data samples and 30 coloums  with additional data coloum as target(malignant or benign) for output comparison

df_cancer = pd.DataFrame(np.c_[breastcancer_df['data'],breastcancer_df['target']],columns=np.append(breastcancer_df['feature_names'],['target']))
```

```python
df_cancer.head()
```

```python
df_cancer.tail()
```

# 2. VISUALIZING DATA


```python
#STEP2 ---------- visualizing data 
#pairplot to plot the correleation in the dataframe
#hue = target
#vars = variables (we can choose which variable we want to plot)
sns.pairplot(df_cancer,hue = 'target',vars = ['mean radius','mean texture','mean area','mean perimeter','mean smoothness'])
```

```python
#countplot to count the sample
sns.countplot(df_cancer['target'])
```

```python
#scatterplot - for plotting specific to variable
sns.scatterplot(x= 'mean area' , y ='mean smoothness' ,hue ='target',data= df_cancer)
```

```python
plt.figure(figsize = (40,20)) #adjusting out figure size
sns.heatmap(df_cancer.corr(),annot=True)
```

# 3. SPLITTING THE DATA INTO TRAINING AND TESTING DATA

```python

X = df_cancer.drop(['target'],axis = 1)
```

```python
X
```

```python
y = df_cancer['target'] #output or target values
```

```python
y
```

```python
X_train ,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=5)
```

```python
X_train
```

```python
X_test
```

# 4. SUPPORT VECTOR MACHINE WITHOUT PARAMETER TUNING

```python
supportvector_model = SVC()
```

```python
supportvector_model.fit(X_train,y_train)
```

## 4.1 EVALUATION

```python

y_predict = supportvector_model.predict(X_test)
y_predict
```

```python
cm = confusion_matrix(y_test,y_predict)
```

```python
sns.heatmap(cm,annot =True)
```

```python
print(classification_report(y_test,y_predict))
```

## 5. FEATURE SCALING - DATA NORMALIZATION (range our data [0,1])

```python

min_train = X_train.min();
range_train = (X_train - min_train).max()
X_train_scaled = (X_train - min_train)/range_train
```

```python
sns.scatterplot(x=X_train_scaled['mean area'],y= X_train_scaled['mean smoothness'],hue = y_train)
```

```python
#feature scaling on testing data set
min_test = X_test.min();
range_test = (X_test - min_test).max()
X_test_scaled = (X_test - min_test)/range_test
```

## 5.1 SVM ON SCALED TRAIN TEST SET

```python
supportvector_model.fit(X_train_scaled,y_train)
```

```python
y_predict = supportvector_model.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test,y_predict)
```

```python
sns.heatmap(cm,annot= True)
```

```python
print(classification_report(y_test,y_predict))
```

# 6. SVM PARAMETER TUNING

```python
#SVM PARAMETER TUNING 

#SKLEARN do this for us and give best c and gamma parameter

parameter_grid = {'C':[0.1,1,10,100],'gamma':[1,0.1,0.01,0.001],'kernel' :['rbf']}
```

```python
from sklearn.model_selection import GridSearchCV #SEARCH WITHIN OUR GRID
#verbose = show how many value to show 
grid = GridSearchCV(SVC(),parameter_grid ,refit = True , verbose= 4)
```

```python
grid.fit(X_train_scaled , y_train)
```

```python
#get our best value of gamma and c
grid.best_params_
```

```python
grid_predictions = grid.predict(X_test_scaled) #predict with best value fof c and gamma and scaled input
```

```python
confusion_metric = confusion_matrix(y_test,grid_predictions)
sns.heatmap(confusion_metric,annot= True)
```

```python
print(classification_report(y_test,grid_predictions))
```
